```python
import matplotlib.pyplot as plt
import random
height, weight, score = [], [], []
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
plt.rc('font',family='Malgun Gothic')
plt.title('키와 몸무게의 상관관계')
plt.scatter(height, weight, s=score, c=score, cmap='rainbow', alpha=0.7)
plt.colorbar(label='체육점수')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.title('오디션 프로그램 득표 현황')
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
plt.plot(singer, week1, label='첫째주', color='hotpink')
plt.plot(singer, week2, label='둘째주', color='royalblue')
plt.legend() #범례
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font',family='Malgun Gothic')
plt.title('오디션 프로그램 득표 현황')

singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

for i in range(len(week1)):
    week1[i] = -week1[i] #두 리스트 中 한 쪽 X축을 음수로 바꿈
    
plt.rcParams['axes.unicode_minus'] = False #그래프 축 값 '-' 기호 활성화
    
plt.barh(singer, week1, label='첫째주', color='hotpink')
plt.barh(singer, week2, label='둘째주', color='royalblue')
plt.legend() #범례
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font',family='Malgun Gothic')
plt.title('오디션 프로그램 득표 현황')

singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

plt.bar(range(1,14,3),week1,label='첫째주',color='hotpink')
plt.bar(range(2,15,3),week2,label='둘째주',color='royalblue')
plt.xticks(range(1,14,3), singer)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

height1, weight1, height2, weight2, score1, score2 = [],[],[],[],[],[]
for i in range(100):
    height1.append(random.randint(140,180))
    weight1.append(random.randint(40,80))
    score1.append(random.randint(0,200))
    height2.append(random.randint(160,200))
    weight2.append(random.randint(50,100))
    score2.append(random.randint(0,200))

plt.rc('font',family='Malgun Gothic')
plt.title('키와 몸무게의 상관관계')
plt.scatter(height1, weight1, s=score1, color='crimson', alpha=0.7, label='그룹1')
plt.scatter(height2, weight2, s=score2, color='indigo', alpha=0.7, label='그룹2')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.legend()
plt.show()
```

```python
#실습1번
import matplotlib.pyplot as plt

weekday = [3,2,16,19,9,12,23,16] #주중 자동차 통행량
weekend = [1,7,13,15,19,10,19,12] #주말 자동차 통행량

plt.rc('font',family='Malgun Gothic')
plt.title('주중/주말 자동차 통행량 비교')
plt.plot(range(0,8),weekday,label='주중 통행량', color='crimson')
plt.plot(range(0,8),weekend, label='주말 통행량', color='royalblue')
plt.legend()
plt.show()
```

```python
#실습2번
import matplotlib.pyplot as plt

weekday = [3,2,16,19,9,12,23,16] #주중 자동차 통행량
weekend = [1,7,13,15,19,10,19,12] #주말 자동차 통행량

for i in range(len(weekday)):
    weekday[i] = -weekday[i]

plt.rc('font',family='Malgun Gothic')
plt.title('주중/주말 자동차 통행량 비교')
plt.barh(range(0,8),weekday,label='주중 통행량', color='crimson')
plt.barh(range(0,8),weekend, label='주말 통행량', color='royalblue')
plt.legend()
plt.show()
```

```python
#실습3번
import matplotlib.pyplot as plt

weekday = [3,2,16,19,9,12,23,16] #주중 자동차 통행량
weekend = [1,7,13,15,19,10,19,12] #주말 자동차 통행량

plt.rc('font',family='Malgun Gothic')
plt.title('주중/주말 자동차 통행량 비교')
plt.boxplot([weekday, weekend], labels=['주중 통행량', '주말 통행량'])
plt.show()
```
