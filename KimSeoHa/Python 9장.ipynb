```python
#그래프로 나타내기
import matplotlib.pyplot as plt 
s_mon=[0,0,0]       #10,11,12의 월별 지출액 저장할 리스트
for row in data:
    if row[-1]=='전표매입':
        mon,payment=int(row[0].split('-')[1]),int(row[-3])  #-으로 구분해서 월을 mon변수에 지출액을 payment에 저장
        idx=mon-10
        s_mon[idx]+=payment  #각각에 월에 해당하는 지출액 저장
        
plt.rc('font',family='Malgun Gothic')
plt.title('10~12월 지출현황')
plt.bar(['10월','11월','12월'],s_mon,color='royalblue')
plt.show()
```

```python
#그래프로 나타내기
import matplotlib.pyplot as plt 
taxi=[0,0,0]    #10,11,12의 월별 지출액 저장할 리스트
for row in data:
    if row[-1]=='전표매입'and'택시'in row[5]:     #전표매입이라고 되어있고 택시라고 되어있으면
        mon,payment=int(row[0].split('-')[1]),int(row[-3])    #-으로 구분해서 두번째에 있는 숫자, row[-3]에 있는 숫자
        idx=mon-10
        taxi[idx]+=payment
        
plt.rc('font',family='Malgun Gothic')
plt.title('10~12월 지출현황')
plt.plot(['10월','11월','12월'],taxi,color='crimson',label='택시비 지출액')
plt.legend()
plt.show()
```

```python
#그래프로 나타내기
import matplotlib.pyplot as plt 
taxi=[0,0,0]    #10,11,12의 월별 지출액 저장할 리스트
deli=[0,0,0]       #10,11,12의 월별 지출액 저장할 리스트
for row in data:
    if row[-1]=='전표매입'and'택시'in row[5]:     #전표매입이라고 되어있고 택시라고 되어있으면
        mon1,payment1=int(row[0].split('-')[1]),int(row[-3])    #-으로 구분해서 두번째에 있는 숫자, row[-3]에 있는 숫자
        idx1=mon1-10
        taxi[idx1]+=payment1
    elif row[-1]=='전표매입'and row[5]=='(주)우아한형제들':
        mon2,payment2=int(row[0].split('-')[1]),int(row[-3])  #-으로 구분해서 월을 mon변수에 지출액을 payment에 저장
        idx2=mon2-10
        deli[idx2]+=payment2  #각각에 월에 해당하는 지출액 저장
        
plt.rc('font',family='Malgun Gothic')
plt.title('10~12월 택시비/배달음식비 지출현황')
plt.plot(['10월','11월','12월'],taxi,color='crimson',label='택시비 지출액')
plt.plot(['10월','11월','12월'],deli,color='indigo',label='배달음식비 지출액')
plt.legend()
plt.show()
```

```python
#지출액 상위 10개 가맹점 골라 그래프 그리기
import operator
import matplotlib.pyplot as plt 

spending={}       #가게명:지출액 형태로 저장하기 위함->딕셔너리를 사용
for row in data:
    if row[-1]=='전표매입':
        store,payment=row[-4],int(row[-3])
        if store not in spending.keys():
            spending[store]=payment
        else:
            spending[store]+=payment
top10=sorted(spending.items(),key=operator.itemgetter(1),reverse=True)[:10] #10개만
top10_store,top10_amount=[],[]
for t in top10:
    top10_store.append(t[0])
    top10_amount.append(t[1])

plt.rc('font',family='Malgun Gothic')
plt.title('10~12월 지출 top10')
plt.barh(top10_store,top10_amount,color='b')
plt.show()
```

```python
#1
import csv

f=open('card.csv',encoding='utf8')   #파일 열기
data=csv.reader(f)  #파일 읽기
next(data)   #헤더 제게(리스트 변환 전에 이루어짐)
data=list(data)  #리스트로 변환

payment=0
for row in data:
    if row[-1]!='전표매입':
        payment+=int(row[-3])
print(payment)
```

```python
#2
import matplotlib.pyplot as plt 
de=[0,0,0]  #10월, 11월, 12월 저장
for row in data:
    if row[-1]=='전표매입':
        payment,store=int(row[-3]),row[-4]
        if '백화점'in store:
            mon=int(row[0].split('-')[1])
            idx=mon-10
            de[idx]+=payment
            
plt.rc('font',family='Malgun Gothic')
plt.title('10~12월 백화점 지출액')
plt.plot(['10월','11월','12월'],de)
plt.show()
```
