```python
#꺾은선 그래프
import matplotlib.pyplot as plt    #matplotlib.pyplot를 plt라고 하겠다.
x=[3,4,5,6,7]
y=[1,5,7,3,7]
plt.plot(x,y)       #.plot->꺾은선 그래프  /  x축,y축
plt.show()          #그래프 보여주는 명령어
```

```python
#그래프 꾸미기
import matplotlib.pyplot as plt    #matplotlib.pyplot를 plt라고 하겠다.
x=['3월','4월','5월','6월','7월']
y=[1,5,7,3,7]
plt.figure(dpi=100)  #해상도
#plt.figure(figsize=(8,4))

plt.rc('font',family='Malgun Gothic') #글꼴
plt.title('월별 판매 실적') #제목

plt.plot(x,y,color='y',label='단위: 개수') #x축, y축, 색, 라벨
plt.legend(loc='lower right') #라벨 위치 변경

plt.xlabel('월')  #x축 이름
plt.ylabel('매출')  #y축 이름
plt.show()
```

```python
#막대 그래프
import matplotlib.pyplot as plt    #matplotlib.pyplot를 plt라고 하겠다.
x=['3월','4월','5월','6월','7월']
y=[1,5,7,3,7]
plt.rc('font',family='Malgun Gothic')
plt.title('월별 판매 실적')

plt.bar(x,y,color='r',label='단위: 개수')   #막대 그래프
plt.legend()

plt.xlabel('월')  #x축 이름
plt.ylabel('매출')  #y축 이름
plt.show()
```

```python
#막대 방향 전환
import matplotlib.pyplot as plt    #matplotlib.pyplot를 plt라고 하겠다.
x=['3월','4월','5월','6월','7월']
y=[1,5,7,3,7]
plt.rc('font',family='Malgun Gothic')
plt.title('월별 판매 실적')

plt.barh(x,y,color='r',label='단위: 개수')   #수평 막대 그래프->기준이 y축
plt.legend()

plt.xlabel('월')  #x축 이름
plt.ylabel('매출')  #y축 이름
plt.show()
```

```python
#산점도 꾸미기
import random
import matplotlib.pyplot as plt    #matplotlib.pyplot를 plt라고 하겠다.

height,weight=[],[] #변수 한번에 정의 가능(리스트도 가능)
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(40,80))
    
plt.rc('font',family='Malgun Gothic')
plt.title('키와 몸무게의 상관관계')
plt.scatter(height,weight,alpha=0.5,c=height,cmap='YlOrRd')  #투명도 지정, 색상의 기준은 키, 컬러맵 지정
plt.colorbar(label='키') #숫자별 색상값 나타내는 컬러바 보여줌
plt.xlabel('키')
plt.ylabel('몸무게')
plt.show()
```

```python
#점크기 조절
import random
import matplotlib.pyplot as plt    #matplotlib.pyplot를 plt라고 하겠다.

height,weight,score=[],[],[] #변수 한번에 정의 가능(리스트도 가능)
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(40,80))
    score.append(random.randint(1,700))
plt.rc('font',family='Malgun Gothic')
plt.title('키와 몸무게, 체육 점수의 상관관계')
plt.scatter(height,weight,s=score,c=score,cmap='rainbow',alpha=0.6)     #점을 score에 따라 조절
plt.xlabel('키')
plt.ylabel('몸무게')
plt.colorbar(label='체육점수')
plt.show()
```

```python
#그래프 겹쳐 그리기
import matplotlib.pyplot as plt
singer=['A','B','C','D','E']
week1=[42,58,19,92,84]
week2=[53,52,48,98,73]
plt.rc('font',family='Malgun Gothic')
plt.title('오디션 프로그램 득표 현황')
plt.plot(singer,week1,label='첫째 주')
plt.plot(singer,week2,label='둘째 주')
plt.legend()  #라벨 표시하기 위한 명령어
plt.show()
```

```python
#가로형 그래프 겹쳐 그리기
import matplotlib.pyplot as plt
singer=['A','B','C','D','E']
week1=[42,58,19,92,84]
week2=[53,52,48,98,73]
for i in range(len(week1)):
    week1[i]=-week1[i]        #week1을 음수로 만들어주기 위함(첫째 주 x축이 음수임)
plt.rcParams['axes.unicode_minus']=False #음수 나타내기 위함

plt.rc('font',family='Malgun Gothic')
plt.title('오디션 프로그램 득표 현황')
plt.barh(singer,week1,label='첫째 주')
plt.barh(singer,week2,label='둘째 주')
plt.legend()  #라벨 표시하기 위한 명령어
plt.show()
```

```python
#세로형 막대그래프 겹쳐 그리기1
import matplotlib.pyplot as plt

week1=[42,58,19,92,84]
week2=[53,52,48,98,73]

plt.rc('font',family='Malgun Gothic')
plt.title('오디션 프로그램 득표 현황')
plt.bar(range(1,6),week1,label='첫째 주')
plt.bar(range(7,12),week2,label='둘째 주')
plt.xticks(range(1,12),['A','B','C','D','E',' ','A','B','C','D','E'])
                
plt.legend()  #라벨 표시하기 위한 명령어
plt.show()
```

```python
#세로형 막대 그래프 겹쳐 그리기 2
import matplotlib.pyplot as plt
singer=['A','B','C','D','E']
week1=[42,58,19,92,84]
week2=[53,52,48,98,73]

plt.rc('font',family='Malgun Gothic')
plt.title('오디션 프로그램 득표 현황')
plt.bar(range(1,14,3),week1,label='첫째 주')
plt.bar(range(2,15,3),week2,label='둘째 주')
plt.xticks(range(1,14,3),singer)
                
plt.legend()  #라벨 표시하기 위한 명령어
plt.show()
```

```python
#산점도 겹쳐 그리기
import random
import matplotlib.pyplot as plt

height1,weight1,height2,weight2=[],[],[],[] #변수 한번에 정의 가능(리스트도 가능)
for i in range(100):
    height1.append(random.randint(100,200))
    weight1.append(random.randint(30,55))
    height2.append(random.randint(150,220))
    weight2.append(random.randint(50,80))
    
plt.rc('font',family='Malgun Gothic')
plt.title('키와 몸무게의 상관관계')

plt.scatter(height1,weight1,color='crimson',alpha=0.7,label='그룹 1')
plt.scatter(height2,weight2,color='indigo',alpha=0.7,label='그룹 2')
plt.legend()

plt.xlabel('키')
plt.ylabel('몸무게')

plt.show()
```
