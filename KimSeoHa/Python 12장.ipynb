```python
#1.필요한 데이터 파일 모두 읽어 들이기
import csv

#인구 데이터 읽어와서 리스트로 저장 
f1=open('LOCAL_PEOPLE_DONG_201912.csv',encoding='utf8')
data=csv.reader(f1)
next(data)
data=list(data)

#행정동코드 데이터 읽어와서 리스트로 저장
f2=open('dong_code.csv')
code_data=csv.reader(f2)
next(code_data)
next(code_data)
code_data=list(code_data)

#위치 정보 데이터 읽어와서 리스트로 저장
f3=open('dong_location.csv')
location_data=csv.reader(f3)
next(location_data)
location_data=list(location_data)

#인구 데이터의 데이터 타입 변환
for row in data:
    for i in range(1,32):
        if i<=2:
            row[i]=int(row[i])
        else:
            row[i]=float(row[i])

#행정동코드 데이터의 데이터 타입 변환
for row in code_data:
    row[0],row[1]=int(row[0]),int(row[1])  #7자리 행정동 코드, 8자리 행정동 코드
    
#위치 정보 데이터의 데이터 타입 변환
for row in location_data:      #row[2]:행정동코드, row[-2]:경도, row[-1]:위도
    row[2],row[-2],row[-1]=int(row[2]),float(row[-2]),float(row[-1])
    
#2.행정동별 인구수를 저장하는 빈 딕셔너리를 만들고 행정동별 총생활인구수를 딕셔너리에 저장하기
dong_population={} #빈딕셔너리 생성

for row in data:
    dong_code,p=row[2],row[3]  #행정동 코드, 총생활인구수 
    if dong_code not in dong_population:
        dong_population[dong_code]=p  #딕셔너리에 없으면 키:값 추가
    else:
         dong_population[dong_code]+=p

#3.인구수를 기준으로 내림차순 정렬, 앞에서 10개 항목만 별도로 저장
import operator

top10=sorted(dong_population.items(),key=operator.itemgetter(1),reverse=True)[:10]
#print(top10)->키;값 쌍으로 출력

top10_code8=[] #행정동 코드 저장할 리스트
for t in top10:   # = top10_code8=[t[0] for t in top10]
    top10_code8.append(t[0])   #행정동 코드 추가  
#리스트 만들기
top10_name=[0 for i in range(10)]  #상위 10개 지역
top10_lat=[0 for i in range(10)]
top10_long=[0 for i in range(10)]

#4.8자리 행정동코드에 해당하는 행정동명 찾기
for i in range(len(top10_code8)):  #행정동코드길이만큼
    for row in code_data:
        if row[1]==top10_code8[i]:   #행정동코드와 top10_code8[i]에 저장된 값이 같으면
            top10_name[i]=row[-1]    # top10_name[i]에 행정동명 저장

#5.행정동명에 해당하는 위도, 경도를 찾아 리스트에 저장
for row in location_data:
    if row[3]==top10_name[i]:
        top10_lat[i],top10_long[i]=row[-1],row[-2]
    
#추가. 행정동명으로 위치 정보를 찾지 못하면 7자리 행정동코드로 찾기
for i in range(len(top10_lat)):
    #위도나 경도가 0인 요소가 있다면
    if top10_lat[i]==0 or top10_long[i]==0:
        #code_data에서 7자리 행정동코드를 구해 dong7에 저장
        for row in code_data:
            if top10_code8[i]==row[1]:
                dong7=row[0]
        #location _data 에서 7자리 행정동코드가 dong7인 행을 찾아 위도와 경도를 구하기
        for row in location_data:
            if dong7==row[2]:
                top10_lat[i],top10_long[i]=row[-1],row[-2]

#6.완성된 리스트 지도 그리기
import folium

#행정동명 앞에 순위 표시하기
for i in range(len(top10_name)):
    top10_name[i]=str(i+1)+'.'+top10_name[i]
    
#지도 그리기
map_top10=folium.Map([top10_lat[0],top10_long[0]],zoom_start=12)
for i in range(10):
    folium.Marker(
        [top10_lat[i],top10_long[i]],tooltip=top10_name[i],icon=folium.Icon(color='red',icon='star')
    ).add_to(map_top10)
map_top10     
```

```python
#추가 분석:여성
import csv

#인구 데이터 읽어와서 리스트로 저장 
f1=open('LOCAL_PEOPLE_DONG_201912.csv',encoding='utf8')
data=csv.reader(f1)
next(data)
data=list(data)

#행정동코드 데이터 읽어와서 리스트로 저장
f2=open('dong_code.csv')
code_data=csv.reader(f2)
next(code_data)
next(code_data)
code_data=list(code_data)

#위치 정보 데이터 읽어와서 리스트로 저장
f3=open('dong_location.csv')
location_data=csv.reader(f3)
next(location_data)
location_data=list(location_data)

#인구 데이터의 데이터 타입 변환
for row in data:
    for i in range(1,32):
        if i<=2:
            row[i]=int(row[i])
        else:
            row[i]=float(row[i])

#행정동코드 데이터의 데이터 타입 변환
for row in code_data:
    row[0],row[1]=int(row[0]),int(row[1])  #7자리 행정동 코드, 8자리 행정동 코드
    
#위치 정보 데이터의 데이터 타입 변환
for row in location_data:      #row[2]:행정동코드, row[-2]:경도, row[-1]:위도
    row[2],row[-2],row[-1]=int(row[2]),float(row[-2]),float(row[-1])
    
#2.행정동별 인구수를 저장하는 빈 딕셔너리를 만들고 행정동별 총생활인구수를 딕셔너리에 저장하기
dong_population={} #빈딕셔너리 생성

for row in data:
    #여성 인구수를 저장하는 변수 p
    dong_code,p=row[2],sum(row[18:32])  
    if dong_code not in dong_population:
        dong_population[dong_code]=p  #딕셔너리에 없으면 키:값 추가
    else:
         dong_population[dong_code]+=p

#3.인구수를 기준으로 내림차순 정렬, 앞에서 10개 항목만 별도로 저장
import operator

top10=sorted(dong_population.items(),key=operator.itemgetter(1),reverse=True)[:10]
#print(top10)->키;값 쌍으로 출력

top10_code8=[] #행정동 코드 저장할 리스트
for t in top10:   # = top10_code8=[t[0] for t in top10]
    top10_code8.append(t[0])   #행정동 코드 추가  
#리스트 만들기
top10_name=[0 for i in range(10)]  #상위 10개 지역
top10_lat=[0 for i in range(10)]
top10_long=[0 for i in range(10)]

#4.8자리 행정동코드에 해당하는 행정동명 찾기
for i in range(len(top10_code8)):  #행정동코드길이만큼
    for row in code_data:
        if row[1]==top10_code8[i]:   #행정동코드와 top10_code8[i]에 저장된 값이 같으면
            top10_name[i]=row[-1]    # top10_name[i]에 행정동명 저장

#5.행정동명에 해당하는 위도, 경도를 찾아 리스트에 저장
for row in location_data:
    if row[3]==top10_name[i]:
        top10_lat[i],top10_long[i]=row[-1],row[-2]
    
#추가. 행정동명으로 위치 정보를 찾지 못하면 7자리 행정동코드로 찾기
for i in range(len(top10_lat)):
    #위도나 경도가 0인 요소가 있다면
    if top10_lat[i]==0 or top10_long[i]==0:
        #code_data에서 7자리 행정동코드를 구해 dong7에 저장
        for row in code_data:
            if top10_code8[i]==row[1]:
                dong7=row[0]
        #location _data 에서 7자리 행정동코드가 dong7인 행을 찾아 위도와 경도를 구하기
        for row in location_data:
            if dong7==row[2]:
                top10_lat[i],top10_long[i]=row[-1],row[-2]

#6.완성된 리스트 지도 그리기
import folium

#행정동명 앞에 순위 표시하기
for i in range(len(top10_name)):
    top10_name[i]=str(i+1)+'.'+top10_name[i]
    
#지도 그리기
map_top10=folium.Map([top10_lat[0],top10_long[0]],zoom_start=12)
for i in range(10):
    #1~5위 지역은 진한 빨간색, 6-7위 지역은 분홍색으로 표시
    if i<5:
        folium.Marker(
            [top10_lat[i],top10_long[i]],tooltip=top10_name[i],icon=folium.Icon(color='red',icon='female',prefix='fa')
        ).add_to(map_top10)
    else:
        folium.Marker(
            [top10_lat[i],top10_long[i]],tooltip=top10_name[i],icon=folium.Icon(color='pink',icon='female',prefix='fa')
        ).add_to(map_top10)
map_top10 
```
