```python
#상품 클래스
class Product:
    def __init__(self,n,p,s):
        self.name=n
        self.price=p
        self.stock=s
    def information(self):
        print('상품이름 :',self.name) #객체 속성 사용-> 객체명.속성
        print('상품가격 :',self.price)
        print('재고수량 :',self.stock)
#회원클래스
class Member:
    def __init__(self,n,j):
        self.name=n
        self.join_date=j
        self.purchase_list=[] #구매내역
        self.purchase_amount=0  #누적 구매금액

    def information(self):
        print('회원이름 :',self.name)
        print('가입날짜 :',self.join_date)
        print('구매내역 :',self.purchase_list)
        print('누적 구매금액 :',self.purchase_amount)
    def buy(self, product, count):
        print(self.name,'고객님이',product.name,count,'개 구매!')
        product.stock-=count         #구매수량만큼 재고 감소
        self.purchase_list.append(product.name)  #구매내역 상품 추가 
        self.purchase_amount+=product.price*count   #누적 구매금액 추가

#물건등록
socks=Product('socks',1000,10)
books=Product('books',17500,15)

#회원가입
Anna=Member('Anna','20200402')  #객체생성(Member 클래스 객체)
Grace=Member('Grace','20200130')

Anna.buy(socks,1)
Anna.information()
print('\n')
socks.information()
print('\n')
Grace.buy(books,2)
Grace.information()
print('\n')

#새로운 물건 입고
pen=Product('pen',1200,25)

Anna.buy(pen,5)
Anna.information()
```

```python
#3,4
#계좌 클래스
class Customer_account:
    def __init__(self,name,bal):
        self.name=name
        self.balance=bal
        print(self.name,'고객님! 계좌를 설계했습니다. 잔고는',self.balance,'원')
        
#은행 클래스
class Bank:
    def __init__(self):    #수수로 속성->초기값이 0으로 설정되어야함
        self.commission=0
        
    #입금 거래,customer에 customer_account 객체 전달
    def deposit(self,customer,amount):
        self.commission+=500
        customer.balance+=(amount-500)
        print(customer.name,'고객님',amount,'원 입금')
        print('수수로 500원 차감 후 입금')
        print('입금 후 잔고는',customer.balance,'원')
        
    #출금 거래, customer에 customer_account 객체 전달
    def withdrawal(self,customer,amount):
        self.commission+=500
        customer.balance-=(amount+500)
        print(customer.name,'고객님',amount,'원 출금')
        print('수수로 500원 차감 후 출금')
        print('출금 후 잔고는',customer.balance,'원')
        
    #계좌이체 거래, sender와 reciever에 customer_account 객체 전달
    def send_money(self,sender,reciever,amount):
        self.commission+=800
        sender.balance-=(amount+800)
        reciever.balance+=amount
        print(sender.name,'고객님',reciever.name,'고객님께',amount,'원 송금')
        print(sender.name,'고객님 잔고에서 수수로 800원 차감')
        print('이체 후 잔고는',sender.name,sender.balance,'원',reciever.name,reciever.balance,'원')
        
        

#객체 생성
ryan=Customer_account('라이언',50000)
prodo=Customer_account('프로도',100000)
bank_a=Bank()

#라이언이 15,000원 출금
bank_a.deposit(ryan,10000)

#프로도가 20000원 입금
bank_a.send_money(prodo,ryan,25000)

#프로도가 라이언에게 25000원 송금
bank_a.withdrawal(ryan,30000)

print('은행의 수수료는 수입은',bank_a.commission,'원')
```
