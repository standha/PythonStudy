# 꺾은 선 그래프

```python
import matplotlib.pyplot as plt
sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']
plt.figure(figsize=(8,5)) #그래프 크기 설정 x,y
plt.rc('font',family='AppleGothic')#글꼴 설정 windows = Malgun Gothic (글꼴이 한글일 경우만 사용)
plt.title('월별 판매 실적')#그래프 제목 넣기
plt.plot(month,sales,color='y',label='단위: 개수')#그래프 색 바꾸기, 범례 넣기
plt.legend(loc='lower right')#범례 표시 및 범례 위치를 9등분해 변경 (default는 왼쪽 상단)
plt.xlabel('월')#x축 이름 넣기 
plt.ylabel('매출')#y축 이름 넣기
plt.show()
```

# 막대 그래프

```python
import matplotlib.pyplot as plt
sales = [1,5,7,3,7]
#month = [3,4,5,6,7]
plt.rc('font',family='AppleGothic')
plt.title('월별 판매 실적')
plt.bar(range(3,8),sales,color='r')#연속적인 데이터 일 경우 range 사용 가능, 막대 그래프는 bar를 사용
plt.show()
```

```python
import matplotlib.pyplot as plt
sales = [1,5,7,3,7]
plt.rc('font',family='AppleGothic')
plt.title('월별 판매 실적')
plt.bar(range(2,11,2),sales,color='r')#짝수 월로 설정 할 경우
plt.show()
```

```python
import matplotlib.pyplot as plt
sales = [1,5,7,3,7]
#month = [3,4,5,6,7]
plt.rc('font',family='AppleGothic')
plt.title('월별 판매 실적')
plt.barh(range(3,8),sales,color='r')#누운 그래프는 barh
plt.show()
```

# 산점도

```python
import matplotlib.pyplot as plt
kor = [80,20,50,20,10,50,60,30,60]
eng = [90,40,60,40,10,30,50,70,90]
plt.rc('font',family='AppleGothic')
plt.scatter(kor,eng,color='r',alpha=0.3)#산점도는 scatter명령어 사용, (kor,eng)의 좌표에 점을 찍음
                              #투명도는 alpha 명령어로 조정 가능
plt.xlabel('국어 점수')
plt.ylabel('영어 점수')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

height, weight = [], []
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
plt.rc('font',family='AppleGothic')
plt.title('키와 몸무게의 상관관계')
plt.scatter(height,weight,c=height,cmap='RdPu')# c옵션으로 컬러맵의 기준 축을 정한 뒤 컬러맵의 이름 지정
plt.colorbar(label='키')#숫자별 색상값을 나타내는 컬러바
plt.xlabel('키')
plt.ylabel('몸무게')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

height, weight, score = [], [], []
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
plt.rc('font',family='AppleGothic')
plt.title('키와 몸무게, 체육 점수의 상관관계')
plt.scatter(height, weight, s=score, c=score, cmap='rainbow',alpha=0.5)#s옵션에 score를 추가해 주면 점의 크기 설정 가능
plt.xlabel('키')
plt.ylabel('몸무게')
plt.colorbar(label='체육 점수')
plt.show()
```

# 그래프 겹쳐 그리기

```python
import matplotlib.pyplot as plt

singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
plt.rc('font',family='AppleGothic')
plt.title('오디션 프로그램 득표 현황')
plt.plot(singer,week1,label='첫째 주',color='hotpink')
plt.plot(singer,week2,label='둘째 주',color='royalblue')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
for i in range(len(week1)):
    week1[i] = -week1[i]
plt.rcParams['axes.unicode_minus']=False
plt.rc('font',family='AppleGothic')
plt.title('오디션 프로그램 득표 현황')
plt.barh(singer, week1, label='첫째 주', color='hotpink')
plt.barh(singer, week2, label='둘째 주', color='royalblue')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

plt.bar(range(1,6),week1,label='첫째 주',color='hotpink')
plt.bar(range(7,12),week2,label='둘째 주',color='royalblue')
plt.xticks(range(1,12),['A','B','C','D','E',' ','A','B','C','D','E'])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

height1, weight1, score1, height2, weight2, score2 = [], [], [], [], [], []
for i in range(100):
    height1.append(random.randint(140,180))
    weight1.append(random.randint(40,80))
    score1.append(random.randint(0,200))
    height2.append(random.randint(160,200))
    weight2.append(random.randint(50,100))
    score2.append(random.randint(0,200))
plt.rc('font',family='AppleGothic')
plt.title('키와 몸무게의 상관 관계')
plt.scatter(height1, weight1, score1, color='crimson',alpha=0.7,label='그룹1')#s를 굳이 추가하지 않아도 됨
plt.scatter(height2, weight2, score2, color='indigo',alpha=0.7,label='그룹2')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.legend()
plt.show()
```
